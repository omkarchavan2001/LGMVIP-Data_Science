# # Data Visualization

```python
year = data['iyear'].unique()
years_count = data['iyear'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(15,7))
sns.barplot(x = year, y = years_count, palette = 'tab10')
plt.xticks(rotation=50)
plt.xlabel('Attacking Year', fontsize=18)
plt.ylabel('Number of Attacks Every Year', fontsize=18)
plt.title('Attacks In Years', fontsize=25)
plt.show()
```

```python
Data = data[['iyear','nkill']].groupby(['iyear']).sum()
fig, ax4 = plt.subplots(figsize=(15,7))
Data.plot(kind='bar',color='blue', ax=ax4)
plt.title('People Died Due to Attack', fontsize=25)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Number of killed people', fontsize=18)
```

```python
data['gname'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar', color='red', figsize=(20,10))
plt.title("Top 10 Terrorist group", fontsize=25, color='Green' )
plt.xlabel('Terrorist group name', fontsize=22, color='Green')
plt.ylabel('Attack number', fontsize=22, color='Green')
plt.show()
```

```python
data['country_txt'].value_counts().to_frame().sort_values('country_txt', axis=0, ascending=False).head(10).plot(kind='bar', color='yellow', figsize=(20,10))
plt.title("Top 10 Attacked Countries", fontsize=25, color='Green' )
plt.xlabel('Country Name', fontsize=22, color='Green')
plt.ylabel('Attack number', fontsize=22, color='Green')
plt.show()
```

```python
data['attacktype1_txt'].value_counts().plot(kind='bar', color='purple', figsize=(20,10))
plt.xticks(rotation=50)
plt.xlabel('Attack Type', fontsize=22)
plt.ylabel('Number of attack', fontsize=22)
plt.title('Name of attacktype', fontsize=25)
plt.show()
```
